```python

from intuitlib.client import AuthClient
from intuitlib.enums import Scopes
from google.colab import userdata
from ipywidgets import Button
from IPython.display import display
from IPython.display import HTML

# Initialize AuthClient with your credentials
auth_client = AuthClient(
    client_id=userdata.get('Int_Client_ID'),
    client_secret=userdata.get('Int_Client_Secret'),
    environment="sandbox",  # or "production"
    redirect_uri=userdata.get('Int_Redirect_Uri')
)
auth_url = auth_client.get_authorization_url([Scopes.ACCOUNTING])
display(HTML('<a href="%s">Click here to authorize Intuit</a>'%auth_url))
auth_code=input(prompt='Please enter auth code')
realm_id=input(prompt='Please enter company ID')

auth_client.get_bearer_token(auth_code)

# Access tokens
print("Access Token:", auth_client.access_token)
print("Refresh Token:", auth_client.refresh_token)
#print("Realm_Id:", auth_client.realm_id)
access_token = auth_client.access_token
refresh_token = auth_client.refresh_token
#realm_id = '9341455712468467'

```

```python

from quickbooks import QuickBooks


# Initialize QuickBooks client

client = QuickBooks(
    auth_client=auth_client,
    refresh_token=refresh_token,
    access_token=access_token,
    company_id=realm_id,
    minor_version=75
)

```

```python
from quickbooks.objects.customer import Customer
customer = Customer()
customer.DisplayName = "Vadim"
customer.CompanyName = "Test Company"
customer.save(qb=client)
```

```python

from quickbooks import QuickBooks
from quickbooks.objects.salesreceipt import SalesReceipt
from quickbooks.objects.detailline import SalesItemLine, SalesItemLineDetail
from quickbooks.objects.base import Ref


# Create SalesReceipt
sales_receipt = SalesReceipt()

# Fake CustomerRef (use any valid ID or placeholder)
sales_receipt.CustomerRef = Ref()
sales_receipt.CustomerRef.value = "123"       # Fake customer ID
sales_receipt.CustomerRef.name = "Test Customer"

# Create a fake line item
line = SalesItemLine()
line.Amount = 100
line.Description = "Fake Product"

# Add SalesItemLineDetail with fake ItemRef
line.SalesItemLineDetail = SalesItemLineDetail()
line.SalesItemLineDetail.ItemRef = Ref()
#line.SalesItemLineDetail.ItemRef.value = "50"
line.SalesItemLineDetail.UnitPrice = 50
line.SalesItemLineDetail.ItemRef.name = "Test Item"
line.SalesItemLineDetail.Qty = 2

# Add line to receipt
sales_receipt.Line.append(line)

# Save to QuickBooks
sales_receipt.save(qb=client)

print(f"Sales Receipt created with ID: {sales_receipt.Id}")
```
